```python
import pandas as pd
import copy
```

```python
FILE = "/Users/wuchh/Documents/DETOPT/real_data/demo/demo_inputs/111e6e61e1_stable.node_freqs"
```

```python
node_freq = pd.read_csv(FILE, sep="\t").set_index("cell_id_x_mut_id")
orig_node_freq = copy.deepcopy(node_freq)
```

```python
PAR_VEC = "/Users/wuchh/Documents/DETOPT/real_data/demo/demo_inputs/111e6e61e1_stable.tree"
```

```python
tree = pd.read_csv(PAR_VEC, sep="\t").set_index("NODE_ID")["PARENT_ID"].to_dict()
del tree['ROOT']
```

```python
while tree != {}:
    r = []
    for n in tree.keys():
        if n not in tree.values():
            r.append(n)
            if tree[n] == "ROOT": continue
            #print(n, tree[n])
            p = int(tree[n])
            node_freq.loc[p,:] = node_freq.loc[p,:] + node_freq.loc[int(n),:] 

    for s in r:        
        del tree[s]
    #print("\n")
```
